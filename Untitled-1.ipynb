```python
#Questão 1
import sympy as sp

t, s = sp.symbols('t, s')
a = sp.symbols('a', real=True, positive=True)

x1 = sp.Heaviside(t - 1)
    
x2 = sp.cos(t - (3*sp.pi)/4)*sp.DiracDelta(t-sp.pi)

XS1 = sp.laplace_transform(x1, t, s, noconds=True)

XS2 = sp.laplace_transform(x2, t, s, noconds = True)

XS = XS1 + XS2

XS

```
